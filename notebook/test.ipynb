```python
import os
from dotenv import load_dotenv
from llama_index.llms.openai import OpenAI
from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.core import Settings


load_dotenv()

OPENAI_API_KEY = os.getenv('OPENAI_API_KEY')
OPENAI_MODEL = os.getenv('OPENAI_MODEL')
OPENAI_EMBED_MODEL = os.getenv('OPENAI_EMBED_MODEL')
TEMPERATURE_MODEL = os.getenv('TEMPERATURE_MODEL')
```

```python
llm = OpenAI(
    api_key=OPENAI_API_KEY,
    model=OPENAI_MODEL,
    temperature=TEMPERATURE_MODEL
)
embed_model = OpenAIEmbedding(
    api_key=OPENAI_API_KEY,
    model=OPENAI_EMBED_MODEL
)
```

```python
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
ans = llm.complete("tell me a joke")
```

```python
print(ans)
```

```python
embed = embed_model.get_text_embedding("tell me a joke")
```

```python
print(embed)
```

```python
import requests
import os

# URL of the file to be downloaded
url = "https://res.cloudinary.com/djlo6r396/raw/upload/v1722415798/uit-admin-bot/xlsx/Key_yqlf9z.xlsx"

# Define the directory where the file will be stored
directory = "D:/UIT/DSC-2024/DSC2024/notebook/data"
# Ensure the directory exists
os.makedirs(directory, exist_ok=True)

# Define the path to save the file
file_path = os.path.join(directory, "Key_yqlf9z.xlsx")

# Download the file
response = requests.get(url)
response.raise_for_status()  # Raise an error for bad status codes

# Write the file to the specified path
with open(file_path, "wb") as file:
    file.write(response.content)

print(f"File downloaded and saved to {file_path}")
```
